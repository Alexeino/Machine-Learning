```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
```

```python
sales_df = pd.read_csv("data/Advertising And Sales.csv")
sales_df.head()
```

#### Let's use just single feature to train and fit the model : Newspaper

```python
X_news = sales_df["Newspaper"].values
y_sales = sales_df["Sales"].values
print(X_news.shape)
y_sales
```

```python
# sklearn requires the shape to be 2D : (200,1)
X_news = X_news.reshape(-1,1)
X_news.shape
```

```python
reg = LinearRegression()
reg.fit(X_news,y_sales)
```

```python
predictions = reg.predict(X_news)
```

#### Let's plot the Predicitons in relation to X_news feature

```python
plt.scatter(X_news,y_sales,color="blue")
plt.plot(X_news,predictions,color="red")
plt.xlabel("News Expenditure")
plt.ylabel("Sales")
plt.show()
```
